### Research Question 1 (What is the perfect time of a movie ? )

```python
high_profit = df[df["profit"] > df["profit"].mean()]
high_avg_vote = df[df["runtime"] > df["runtime"].mean()]
```

```python
df["runtime"].describe()
```

```python
average_perfect_time = high_profit["runtime"].mean()
average_perfect_time
```

```python
average_perfect_time2 = high_avg_vote["runtime"].mean()
average_perfect_time2
```

```python
df.plot.scatter(x="runtime", y="profit")
```

# Conclusion1

>So the average runtime of the most successful movies is between 113 and 123 minutes although the best movie (in revenue) is avatar whose runtime is 162 but it is not standard. Personally, I get bored with movies that are longer than 2 hours.

## Q2: What genres are most popular?

```python
greater_avg_profit = df[df["profit"] > df["profit"].mean()]
```

```python
genres = high_profit["genres"]
genres
```

```python
all_genres = df["genres"].str.cat(sep = '|')
all_genres = pd.Series(all_genres.split('|'))
count = all_genres.value_counts()
count
```

```python
count.plot(kind="pie", fontsize = 12, figsize=(12, 15))
```

```python
high_profit_genres = high_profit["genres"].str.cat(sep = '|')
high_profit_genres = pd.Series(high_profit_genres.split('|'))
high_profit_count = high_profit_genres.value_counts()
high_profit_count
```

```python
high_profit_count.plot.bar(color="steelblue", fontsize = 12, figsize=(8, 8))
```
